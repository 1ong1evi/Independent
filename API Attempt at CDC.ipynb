```python
# import libraries 
from requests import Request, Session
import json
import pandas as pd
```

```python
# using api from Census data 
url = 'https://api.census.gov/data/2021/cbp?get=NAME,EMP&for=state:*'
```

```python
# renaming session to be easier to be called
session = Session()
response = session.get(url)
# loading data as json text
data = json.loads(response.text)
print(data)
```

```python
# the data type in this case is a list
type(data)
```

```python
# creating the dataframe with columnns reprsenting the data we got back
df = pd.DataFrame(data, columns=['NAME','EMP','state'])
df
```

```python
# removing the first index where it was previously the column categories
df.drop(0, inplace = True)
df
```

```python
# changing the data of EMP to a int because it wasn't plotting the data due to it not being numeric
df['EMP']=df['EMP'].astype(int)
df.dtypes
# regarding state it can and should be converted to numeric incase I want to call it as such
```

```python
df['EMP'].plot.bar()
# plotting the values
df.sort_values(by='EMP', ascending=False)
```

```python
# checking max value for EMP 
df['EMP'].max()
```
