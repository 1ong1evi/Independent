```python
from requests import Request, Session
import json
import pandas as pd
session = Session()
```

```python
url = 'https://api.census.gov/data/2021/cbp?get=NAME,EMP&for=state:*'
```

```python
session = Session()
response = session.get(url)
data = json.loads(response.text)
print(data)
```

```python
type(data)
```

```python
df = pd.DataFrame(data, columns=['NAME','EMP','state'])
df
```

```python
df = df.set_index('NAME')
df
```

```python
df.drop('NAME', inplace = True)
df
```
